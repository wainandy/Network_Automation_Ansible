**NETWORK AUTOMATION USING ANSIBLE**

Ansible is a open-source network automation tool that handles configuration management, software provisioning and application deployment. These are the following components and technical terms that make up Ansible:

1. Control Node: It is the device running Ansible tool.e.g. Google Colab, Ubuntu WSL.
2. Managed Node: It is the device that is controlled by Ansible e.g. Cisco IOS XE catalyst 9000 router.
3. Inventory: It is the list of nodes to be managed.
4. Groups: Organise the related nodes into a single category for ease of management and configuration e.g. Routers, Switches.
5. Playbooks: These are simple YAML files that are instructions telling Ansible which tasks to execute on the managed nodes.
6. Modules: Are the specific tools that perform the intended tasks e.g. cisco.ios.ios_command - Runs `show` commands, cisco.ios.ios_config - Configure router, ios_facts- Collect facts, template - generate configs, copy- copy files
7. Variables - Store the data.
8. Jinja2 templates are configuration blueprints by customizing configuration parameters, generate files dynamically and manipulate data on the control nodes. Processed using `ansible.builtin.template` module. Has three delimiter types: `{{ }}` - Used for printing variable values or the expression results into the file, `{%....%}` - Used for logic blocks such as loops, conditional statements.
9. Facts - Are information collected from devices e.g. Hostname, model type, serial number, IOS version
10.  Collections - Adds new capabilities that Ansible can be familiar about e.g. `ansible-galaxy collection install cisco.ios`. Ansible is able to know about Cisco IOS.
11. Registered variables: It allows saving of the results for later use.
12. Tasks are the individual actions performed on the monitored node e.g. show version
13. Rendering is the automated process of combining a blueprint with variables to produce a single, finalized file and sends to the specified destination path.
14. Hosts variable file- It stores custom configurations unique to a specific device or host.

**Simple Ansible Automation Workflow**:

Variables + Jinja2 template = Ansible Output

**Example**

NTP configuration:

Variable (YAML):

`ntp_servers:
    - 10.10.1.11
    - 10.10.1.12`

Jinja2 Template:

`{% for server in ntp_servers %}
ntp server {{ server }}
{% endfor %}`

Ansible will finally generate:

`ntp server 10.10.1.11
 ntp server 10.10.1.12`

* 'Show Clock' Playbook (First Playbook)

````python
%%writefile playbooks/00_preflight.yml
---
- hosts: iosxe
  gather_facts: false

  tasks:

  - name: Show clock
    cisco.ios.ios_command:
      commands:
        - show clock

    register: clock_output

  - debug:
      var: clock_output.stdout_lines
````

* Rendering LoopBack configurations

````python
%%writefile playbooks/04_render.yml
---
- hosts: iosxe
  gather_facts: false
  vars_files:
    - ../group_vars/iosxe.yml

  tasks:

  - template:
      src: ../templates/loopbacks.j2
      dest: rendered/loopbacks.cfg

    delegate_to: localhost
````

### Diff Mode

`Diff` mode allows you to see the exact changes that Ansible *would* make to a device's configuration without actually applying them. It compares the current state of the device with the desired state defined in your playbook and shows the differences.

To run a playbook in `diff` mode, add the `--diff` (or `-D`) flag:

```bash
!ansible-playbook playbooks/your_playbook.yml --diff
```

### Check Mode

`Check` mode (also known as 'dry run' mode) allows Ansible to run a playbook and report what changes *would* be made, but without actually implementing any modifications on the managed nodes. This is useful for verifying playbook logic and syntax.

To run a playbook in `check` mode, add the `--check` (or `-C`) flag:

```bash
!ansible-playbook playbooks/your_playbook.yml --check
```

### Backup

Many Ansible network modules support a `backup` option, which automatically creates a backup of the device's configuration *before* any changes are applied. This provides a rollback mechanism in case of unexpected issues.

As seen in the `03_backup.yml` playbook, setting `backup: true` within a task will create a timestamped backup file.

```yaml
- name: Backup running configuration
  cisco.ios.ios_config:
    backup: true
```

### Explicit Changes

Ansible tasks clearly indicate whether a change was made (`changed=true`) or if the device was already in the desired state (`changed=false`). This explicit reporting helps in understanding the impact of each task and ensures that only necessary modifications are applied.

When you run an Ansible playbook, the `PLAY RECAP` section summarizes the changes:

```bash
PLAY RECAP *********************************************************************
sandbox_iosxe              : ok=1   changed=1    unreachable=0    failed=0    skipped=0    rescued=0    ignored=0
```

Here, `changed=1` indicates that one task resulted in a configuration change on `sandbox_iosxe`.
